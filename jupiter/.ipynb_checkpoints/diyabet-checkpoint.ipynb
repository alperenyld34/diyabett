```python
from sklearn.model_selection import train_test_split

x = df.drop(["Sonuc"], axis = 1)
y = df["Sonuc"]

# X verileri, y hedef değişkenleri
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(x_train.shape)
print(x_test.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score, GridSearchCV

# Naive Bayes modelinin oluşturulması
nb_model = GaussianNB()

# Grid Search ile en iyi parametreleri bulma
param_grid = {'var_smoothing': [1e-9, 1e-8, 1e-7]}
grid_search = GridSearchCV(nb_model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

# En iyi parametreleri ve en iyi tahmin doğruluğunu yazdırma
print("En İyi Parametreler :", grid_search.best_params_)
print("En İyi Tahmin Doğruluğu :", grid_search.best_score_)

# En iyi modeli seçme ve eğitimi
best_model = grid_search.best_estimator_
best_model.fit(x_train, y_train)

# Test verileri üzerinde tahmin yapma
y_pred = best_model.predict(x_test)

# Olasılık tahminlerini almak için predict_proba yöntemini kullanma
y_pred_proba = best_model.predict_proba(x_test)[:, 1]

# Performans metriklerinin hesaplanması
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred_proba)

print("Tahmin Doğruluğu :", accuracy)
print("Hassasiyet :", precision)
print("Duyarlılık :", recall)
print("F1 Skoru :", f1)
print("AUC Değeri :", auc)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest modelinin oluşturulması
rf_model = RandomForestClassifier()

# Grid Search Cross Validation için parametrelerin belirlenmesi
param_grid = {'n_estimators': [100, 200, 500],
              'max_depth': [3, 5, 7],
              'min_samples_split': [2, 5, 10]}

# Grid Search Cross Validation ile en iyi parametreleri bulma
grid_search = GridSearchCV(rf_model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

# En iyi parametreleri ve en iyi tahmin doğruluğunu yazdırma
print("En İyi Parametreler :", grid_search.best_params_)
print("En İyi Tahmin Doğruluğu :", grid_search.best_score_)

# En iyi modeli seçme ve eğitimi
best_model = grid_search.best_estimator_
best_model.fit(x_train, y_train)

# Test verileri üzerinde tahmin yapma
y_pred = best_model.predict(x_test)

# Performans metriklerinin hesaplanması
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

# Özellik önemlerini elde etme
feature_importances = best_model.feature_importances_

print("Tahmin Doğruluğu :", accuracy)
print("Hassasiyet :", precision)
print("Duyarlılık :", recall)
print("F1 Skoru :", f1)
print("AUC Değeri :", auc)
print("Özellik Önemleri :", feature_importances)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Karar Ağacı modelinin oluşturulması
dt_model = DecisionTreeClassifier()

# Grid Search Cross Validation için parametrelerin belirlenmesi
param_grid = {'max_depth': [3, 5, 7],
              'min_samples_split': [2, 5, 10]}

# Grid Search Cross Validation ile en iyi parametreleri bulma
grid_search = GridSearchCV(dt_model, param_grid, cv=10, scoring='accuracy')
grid_search.fit(x_train, y_train)

# En iyi parametreleri ve en iyi tahmin doğruluğunu yazdırma
print("En İyi Parametreler :", grid_search.best_params_)
print("En İyi Tahmin Doğruluğu :", grid_search.best_score_)

# En iyi modeli seçme ve  eğitimi
best_model = grid_search.best_estimator_
best_model.fit(x_train, y_train)

# Test verileri üzerinde tahmin yapma
y_pred = best_model.predict(x_test)

# Performans metriklerinin hesaplanması
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

print("Tahmin Doğruluğu :", accuracy)
print("Hassasiyet :", precision)
print("Duyarlılık :", recall)
print("F1 Skoru :", f1)
print("AUC Değeri :", auc)
```

```python
from sklearn.linear_model import LogisticRegression

# Lojistik Regresyon modelinin oluşturulması
lr_model = LogisticRegression(solver='liblinear')

# Hiperparametrelerin tanımlanması
param_grid = {'C': [0.1, 1, 10, 100], 'penalty': ['l1', 'l2']}

# Grid Search Cross Validation ile en iyi parametreleri bulma
grid_search = GridSearchCV(lr_model, param_grid, cv=2, scoring='accuracy')
grid_search.fit(x_train, y_train)

# En iyi parametreleri ve en iyi tahmin doğruluğunu yazdırma
print("En İyi Parametreler :", grid_search.best_params_)
print("En İyi Tahmin Doğruluğu :", grid_search.best_score_)

# En iyi modeli seçme ve  eğitimi
best_model = grid_search.best_estimator_
best_model.fit(x_train, y_train)

# Test verileri üzerinde tahmin yapma
y_pred = best_model.predict(x_test)

# Performans metriklerinin hesaplanması
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

print("Tahmin Doğruluğu (Test Seti):", accuracy)
print("Hassasiyet :", precision)
print("Duyarlılık :", recall)
print("F1 Skoru :", f1)
print("AUC Değeri :", auc)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# k-En Yakın Komşuluk modelinin oluşturulması
knn_model = KNeighborsClassifier()

# Grid Search ile en iyi parametreleri bulma
param_grid = {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}
grid_search = GridSearchCV(knn_model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

# En iyi parametreleri ve en iyi tahmin doğruluğunu yazdırma
print("En İyi Parametreler :", grid_search.best_params_)
print("En İyi Tahmin Doğruluğu :", grid_search.best_score_)

# En iyi modeli seçme ve eğitimi
best_model = grid_search.best_estimator_
best_model.fit(x_train, y_train)

# Test verileri üzerinde tahmin yapma
y_pred = best_model.predict(x_test)

# Performans metriklerinin hesaplanması
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

print("Tahmin Doğruluğu (Test Seti):", accuracy)
print("Hassasiyet :", precision)
print("Duyarlılık :", recall)
print("F1 Skoru :", f1)
print("AUC Değeri :", auc)
```

```python
from sklearn.svm import SVC

# SVM modelinin oluşturulması
svm_model = SVC(kernel='rbf', C=1.0, random_state=42)

# Çapraz doğrulama ile tahmin doğruluğunun hesaplanması
scores = cross_val_score(svm_model, x_train, y_train, cv=5)
average_accuracy = scores.mean()

print("Ortalama Tahmin Doğruluğu :", average_accuracy)

# Modelin tam eğitimi ve test verileri üzerinde tahmin yapma
svm_model.fit(x_train, y_train)
y_pred = svm_model.predict(x_test)

# Performans metriklerinin hesaplanması
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

print("Tahmin Doğruluğu (Test Seti):", accuracy)
print("Hassasiyet :", precision)
print("Duyarlılık :", recall)
print("F1 Skoru :", f1)
print("AUC Değeri :", auc)
```

```python
import tkinter as tk
from tkinter import messagebox
from sklearn.naive_bayes import GaussianNB
import pandas as pd

# Veri setini yükleme
df = pd.read_csv('diyabet.csv')

# Naive Bayes modelinin oluşturulması ve eğitimi
x = df.drop(["Sonuc"], axis=1)
y = df["Sonuc"]
nb_model = GaussianNB()
nb_model.fit(x, y)

# Arayüz penceresini oluşturma
window = tk.Tk()
window.title("Diyabet Tahmini")
window.geometry("400x700")
window.configure(background="#ffffff")

# Başlık etiketi
title_label = tk.Label(window, text="Diyabet Tahmini", font=("Arial", 20, "bold"), background="#007bff", fg="#ffffff", pady=20)
title_label.pack(fill=tk.X)

# Etiketler
labels = ["Gebelik:", "Glikoz:", "Kan Basıncı:", "Cilt Kalınlığı:",
          "İnsülin:", "BMI:", "Diyabet Soyağacı İşlevi:", "Yaş:"]

# Girdi alanları
entries = []
for label in labels:
    label_frame = tk.Frame(window, background="#ffffff")
    label_frame.pack(pady=10)
    tk.Label(label_frame, text=label, font=("Arial", 14), background="#ffffff").pack(side=tk.LEFT)
    entry = tk.Entry(label_frame, font=("Arial", 14), width=10)
    entry.pack(side=tk.RIGHT)
    entries.append(entry)

# Tahmin sonucu etiketi
result_label = tk.Label(window, text="", font=("Arial", 16, "bold"), background="#ffffff")
result_label.pack(pady=30)

# Tahmin işlemini gerçekleştiren fonksiyon
def predict_diabetes():
    # Kullanıcıdan girilen değerleri alıp numpy dizisine dönüştürme
    input_data = [float(entry.get()) for entry in entries]
    input_array = pd.DataFrame([input_data], columns=x.columns)

    # Tahmin yapma
    result = nb_model.predict(input_array)

    # Sonucu kullanıcıya bildirme
    if result[0] == 0:
        result_label.config(text="Kişi Diyabetli Değil", fg="#007bff")
    else:
        result_label.config(text="Kişi Diyabetli", fg="#ff3b30")

# Tahmin yap butonu
predict_button = tk.Button(window, text="Tahmin Yap", command=predict_diabetes, font=("Arial", 16), bg="#007bff", fg="#ffffff", relief=tk.RAISED, pady=10)
predict_button.pack(fill=tk.X, padx=50, pady=20)

# Arayüzü çalıştırma
window.mainloop()
```
