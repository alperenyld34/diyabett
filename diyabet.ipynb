```python
from sklearn.model_selection import train_test_split

x = df.drop(["Sonuc"], axis = 1)
y = df["Sonuc"]

# X verileri, y hedef değişkenleri
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(x_train.shape)
print(x_test.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import GridSearchCV
import pickle

# Naive Bayes modelinin oluşturulması
nb_model = GaussianNB()

# Grid Search ile en iyi parametreleri bulma
param_grid = {'var_smoothing': [1e-9, 1e-8, 1e-7]}
grid_search = GridSearchCV(nb_model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

# En iyi parametreleri ve en iyi tahmin doğruluğunu yazdırma
print("En İyi Parametreler:", grid_search.best_params_)
print("En İyi Tahmin Doğruluğu:", grid_search.best_score_)

# En iyi modeli seçme ve eğitimi
best_model = grid_search.best_estimator_
best_model.fit(x_train, y_train)

# Test verileri üzerinde tahmin yapma
y_pred = best_model.predict(x_test)

# Olasılık tahminlerini almak için predict_proba yöntemini kullanma
y_pred_proba = best_model.predict_proba(x_test)[:, 1]

# Performans metriklerinin hesaplanması
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred_proba)

print("Tahmin Doğruluğu:", accuracy)
print("Hassasiyet:", precision)
print("Duyarlılık:", recall)
print("F1 Skoru:", f1)
print("AUC Değeri:", auc)

# Modeli .pkl dosyası olarak kaydetme
with open('egitilmis_model.pkl', 'wb') as file:
    pickle.dump(best_model, file)
```

```python
import tkinter as tk
from tkinter import messagebox
import pickle
import matplotlib.pyplot as plt

# Arayüz penceresini oluşturulması
window = tk.Tk()
window.title("Diyabet Tahmin")
window.geometry("400x750")
window.configure(background="#ffffff")

# Başlık
title_label = tk.Label(window, text="Diyabet Tahmini", font=("Cofosans", 18, "bold"), background="#008b8b", fg="#ffffff", pady=25)
title_label.pack(fill=tk.X)

# Değişkenler
labels = ["Gebelik:", "Glikoz:", "Kan Basıncı:", "Cilt Kalınlığı:",
          "İnsülin:", "BMI:", "Diyabet Soyağacı İşlevi:", "Yaş:"]
tooltips = ["Gebe kalınan sayı.\n 0-50 aralığında değer girilebilir.", "2 saatlik oral glikoz tolerans testindeki plazma glikoz konsantrasyonu. \n 0-300 aralığında değer girilebilir.", 
            "Diastolik kan basıncı (mm Hg). \n 0-200 aralığında değer girilebilir. ", "Triceps cilt kıvrım kalınlığı (mm). \n 0-100 aralığında değer girilebilir.",
            " 2 saatlik serum insülini (mu U/ml). \n 0-1000 aralığında değer girilebilir.", "Vücut Kitle İndeksi: Ağırlık (kg) / boyun karesi (m^2). \n 0-100 aralığında değer girilebilir.", 
            "Hastanın diyabet soyağacı ile ilişkilendirilme olasılığı. \n 0-5 aralığında değer girilebilir.", "Yaş. \n 0-150 aralığında değer girilebilir."]
# Değişkenler için belirlenen değer aralıkları
ranges = [(0, 50), (0, 300), (0, 200), (0, 100), 
          (0, 1000), (0, 100), (0, 5), (0, 150)]

# Tooltip sınıfı oluşturma
class ToolTip(object):
    def __init__(self, widget, text='Widget info'):
        self.waittime = 500     #miliseconds
        self.wraplength = 180   #pixels
        self.widget = widget
        self.text = text
        self.widget.bind("<Enter>", self.onEnter)
        self.widget.bind("<Leave>", self.onLeave)
        self.tooltipwindow = None
    def onEnter(self, event=None):
        self.schedule = self.widget.after(self.waittime, self.showTooltip)
    def onLeave(self, event=None):
        self.widget.after_cancel(self.schedule)
        self.hideTooltip()
    def showTooltip(self):
        x, y, _, _ = self.widget.bbox("insert")
        x += self.widget.winfo_rootx() + 25
        y += self.widget.winfo_rooty() + 20
        self.tooltipwindow = tw = tk.Toplevel(self.widget)
        tw.wm_overrideredirect(True)
        tw.wm_geometry(f"+{x}+{y}")
        label = tk.Label(tw, text=self.text, wraplength=self.wraplength, background="#ffffff")
        label.pack()
    def hideTooltip(self):
        tw = self.tooltipwindow
        self.tooltipwindow = None
        if tw:
            tw.destroy()

# Girdi alanları
entries = []
for label, tooltip in zip(labels, tooltips):
    label_frame = tk.Frame(window, background="#ffffff")
    label_frame.pack(pady=10)
    tk.Label(label_frame, text=label, font=("Cofosans", 13), background="#ffffff").pack(side=tk.LEFT)
    entry = tk.Entry(label_frame, font=("Cofosans", 13), width=10, justify='center')
    entry.pack(side=tk.LEFT)
    entries.append(entry)
    info_label = tk.Label(label_frame, text="ℹ", font=("Arial", 13), foreground="#0000ff")
    info_label.pack(side=tk.RIGHT)
    ToolTip(info_label, text=tooltip)

# Tahmin sonuc kısmı
result_label = tk.Label(window, text="", font=("Cofosans", 16, "bold"), background="#ffffff")
result_label.pack(pady=30)

# Modeli yükleme
with open('egitilmis_model.pkl', 'rb') as file:
    best_model = pickle.load(file)

# Tahmin işlemini gerçekleştiren fonksiyon
def predict_diabetes():
    # Kullanıcıdan girilen değerleri alıp numpy dizisine dönüştürme
    input_data = [float(entry.get()) if entry.get().strip() != "" else None for entry in entries]
    
    # Geçerlilik kontrolü
    if None in input_data:
        messagebox.showwarning("Hata", "Geçerli bir değer girin.")
        return

    # Aralık kontrolü
    for data, valid_range in zip(input_data, ranges):
        if not valid_range[0] <= data <= valid_range[1]:
            messagebox.showwarning("Hata", f"{labels[input_data.index(data)]} girişi belirlenen aralığın dışında. Girilmesi gereken aralık: {valid_range[0]}-{valid_range[1]}")
            return

    input_array = pd.DataFrame([input_data], columns=x.columns)

    # Tahmin yapma
    result = best_model.predict(input_array)

    # Sonucu kullanıcıya bildirme
    if result[0] == 0:
        result_label.config(text="Kişi Diyabetli Değil", fg="#007bff")
    else:
        result_label.config(text="Kişi Diyabetli", fg="#ff3b30")
        
# Tahmin yap butonu
predict_button = tk.Button(window, text="Tahmin Yap", command=predict_diabetes, font=("Arial", 16, "bold"), bg="#008b8b", fg="#ffffff", relief=tk.RAISED, pady=10)
predict_button.pack(fill=tk.X, padx=50, pady=20)

# Girdi alanlarını temizlemek için bir fonksiyon oluşturma
def clear_entries():
    for entry in entries:
        entry.delete(0, tk.END)
    result_label.config(text="")  # sonuç metnini de temizle

# Temizle butonu
clear_button = tk.Button(window, text="Temizle", command=clear_entries, font=("Arial", 16, "bold"), bg="#008b8b", fg="#ffffff", relief=tk.RAISED, pady=10)
clear_button.pack(fill=tk.X, padx=50, pady=30)

window.mainloop()
```
